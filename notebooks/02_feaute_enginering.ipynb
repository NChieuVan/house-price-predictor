```python
# Feature Engineering

# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv('../data/processed/house_data_cleaned.csv')
df.head()
```

```python
# init data exploration
print("Dataframe shape:", df.shape)
print("\nDataframe info:")
print(df.info())
print("descibe the dataset")
print(df.describe())
```

```python
# Feature Engineering (Human - Readable)

# Feature 1: Age of the house
df['house_age'] = datetime.now().year - df['year_built']

# Feature 2: Price per square foot
df['price_per_sqft'] = df['price'] / df['sqft']

# Feature 3: Ratio of bedrooms to bathrooms
df['bed_bath_ratio'] = df['bedrooms'] / df['bathrooms']

# Handle infinite values in 'bed_bath_ratio' (if bathrooms is zero)
df['bed_bath_ratio'] = df['bed_bath_ratio'].replace([np.inf, -np.inf], np.nan).fillna(0)

# Quick Look at the new features
print("\nDataframe with new features:")
print(df[['house_age', 'price_per_sqft', 'bed_bath_ratio']].describe())
```

```python
# Visulize the new features
fig,axs = plt.subplots(1,3, figsize=(18,5))
sns.histplot(df['house_age'], kde=True, ax=axs[0])
sns.histplot(df['price_per_sqft'], kde=True, ax=axs[1])
sns.histplot(df['bed_bath_ratio'], kde=True, ax=axs[2])
axs[0].set_title('Distribution of House Age')
axs[1].set_title('Distribution of Price per Square Foot')
axs[2].set_title('Distribution of Bedroom to Bathroom Ratio')
plt.tight_layout()
plt.show()
```

```python
# Save Final Engineered Features (Human - Readable)

selected_features = ['price','sqft','bedrooms','location','year_built','condition',
                     'house_age', 'price_per_sqft', 'bed_bath_ratio']

df[selected_features].to_csv('../data/processed/data_scientists_features.csv', index=False)
print("Final engineered features saved to '../data/processed/data_scientists_features.csv'")
```
