```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv('../data/processed/house_data_cleaned.csv')
print("Dataset loaded successfully. Shape:", df.shape)

print("\nDataset info:")
print(df.info())
```

```python
# check missing values
print("\nChecking for missing values...")
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Describe the dataset
print("\nBasic statistics check:")
print(df.describe())
```

```python
# visualize distributions of key features
plt.figure(figsize=(12, 6))
sns.histplot(df['price'], kde=True)
plt.title('Distribution of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatter plot to observe relationship between 'sqft and 'price'
plt.figure(figsize=(10, 8))
sns.scatterplot(x='sqft', y='price', data=df)
plt.title('Scatter Plot of Square Footage vs Price')
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.show()
```

```python
# Bar chart to see how many houses exist in each location
plt.figure(figsize=(12, 8))
sns.countplot(x='location', data=df)
plt.title('Count of Houses by Location')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar chart showing how average house price varies by bedrooms count
plt.figure(figsize=(10, 6))
sns.barplot(x='bedrooms', y='price', data=df)
plt.title('Average House Price by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Price')
plt.show()
```
